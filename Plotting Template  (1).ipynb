# Plotting Template
You can start off your notebook by creating a title and maybe including some information on what you are plotting. 

In this example I am going to create a picture of yellow stars in the sky above blue water. The stars will be points plotted with star-shaped markers and the water will be a series of sine curves all plotted using the same range of x-values.

```python
# --------IMPORTING LIBRARIES--------
import matplotlib.pyplot as plt
import numpy as np
```

```python
# --------FORMATTING THE DISPLAY--------

# Setting the size of the display window
plt.figure(figsize = (10, 7))
# Setting the axes sizes
plt.xlim((-15, 15))
plt.ylim((-15, 15))
# Setting the axes markers 
plt.xticks([-15, -10, -5, 0, 5, 10, 15])
plt.yticks([-15, -10, -5, 0, 5, 10, 15])
# We do not want grid or axes lines so we left out those formatting lines we saw in the manual.

# --------PLOTTING POINTS AND FUNCTIONS--------

# Creating the list of points I'd like to plot.
points = [(-14, 14), (-13, 10), (-11, 13), (-10, 9), (-8, 14), (-6, 10), (-4, 12), (-3, 13), (-1, 10), (0, 14), (2, 12), (3, 13), (5, 10), (6, 9), (8, 14), (10, 12), (12, 14), (14, 10), (-4, 7), (3, 8), (10, 9)]

# Seperating the points into list of x-coordinates and y-coordinates.
pointsX = [-14, -13, -11, -10, -8, -6, -4, -3, -1, 0, 2, 3, 5, 6, 8, 10, 12, 14, -4, 3, 10]
pointsY = [14, 10, 13, 9, 14, 10, 12, 13, 10, 14, 12, 13, 10, 9, 14, 12, 14, 10, 7, 8, 9]

# Plotting the points with yellow stars of size 10.
plt.plot(pointsX, pointsY, 'y*', markersize = 10)


# Defining first sine function.
def sine1(x):
    y = np.sin(x)
    return y

# Creating x-values for the function to be evaluated at.
X = np.linspace(-15, 15, 200)

# Plotting the function with a blue solid line.
plt.plot(X, sine1(X), 'b-')


# Defining the second sine function.
def sine2(x):
    y = np.sin(x + 1) - 3
    return y

# Since this function will be evaluated at the same x-values as the previous, we can use the same
# list 'X' as above. 

# Plotting the function with a blue solid line. 
plt.plot(X, sine2(X), 'b-')


def sine3(x): 
    y = np.sin(x + 2) - 6
    return y

plt.plot(X, sine3(X), 'b-')


def sine4(x):
    y = np.sin(x + 3) - 9
    return y

plt.plot(X, sine4(X), 'b-')


def sine5(x):
    y = np.sin(x + 4) - 12
    return y

plt.plot(X, sine5(X), 'b-')


def sine6(x):
    y = np.sin(x + 5) - 15
    return y

plt.plot(X, sine6(X), 'b-')


# --------ADDING TEXT TO THE PLOT--------
plt.text(8, 11, 'RabbitMath', size = 20)


plt.show()
```
