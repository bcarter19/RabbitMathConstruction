```python
import matplotlib.pyplot as plt
import numpy as np
```

### The Data Points 
We'll first start off by inputting our data points and creating a plot.

```python
# Seperating the points into list of x-coordinates and y-coordinates.
x = [1, 2, 3, 4, 5]
y = [4, 7, 5, 6, 10]

# Setting the axes sizes
plt.xlim((0, 6))
plt.ylim((0, 11))

# Plotting the points with red circles.
plt.plot(x, y, 'ro')

# Add grid lines
plt.grid()
```

### Plotting Lines
Next we need to define a function that will output a line given its slope and y-intercept.

```python
# The inputs are the slope of the line (m), the intercept of the line (b) and the x-value of a point on the line (x). 
# The output is the y-value of the point on the line. 
def line(m, b, x):
    y = m*x + b
    return y
```

Let's plot the data and a line now. 

```python
# Seperating the points into list of x-coordinates and y-coordinates.
x = [1, 2, 3, 4, 5]
y = [4, 7, 5, 6, 10]

# Setting the axes sizes
plt.xlim((0, 6))
plt.ylim((0, 11))

# Plotting the points with red circles.
plt.plot(x, y, 'ro')

# Plotting the line with slope 1 and intercept 4.
# Creating x-values for the function to be evaluated at.
# The np.linspace function generates 200 evenly spaced values between 0 and 6. 
X = np.linspace(0, 6, 200)
# Plotting the line with slope 1 and y-intercept 4 with a blue solid line.
plt.plot(X, line(1, 4, X), 'b-')

# Add grid lines
plt.grid()
```

### Square Errors
'm' and 'b' are ranges of possible slope and intercept values between 0 and 5 and 0 and 8 respectively. The function SQ(m, b) calculates the squared error between the data points and the line with slope m and intercept b. 

```python
# Creating slope values for the squared error function to be evaluated at. 
# The np.linspace() function generates 400 evenly spaced values between 0 and 5. 
m = np.linspace(0, 5, 400)
# Creating intercept values for the squared error function to be evaluated at. 
# The np.linspace() function generates 400 evenly spaced values between 0 and 8. 
b = np.linspace(0, 8, 400)

def SQ(m, b):
    global x, y
    total = 0
    i = 0
    while i < len(x): 
        total = total + (y[i] - line(m, b, x[i]))**2
        i = i + 1
    return total 
```

### Creating the Contour Diagram

```python
M, B = np.meshgrid(m, b)
Z = SQ(M, B)
```

```python
# Setting the size of the display window
plt.figure(figsize = (10, 7))
# Setting the axes sizes
plt.xlim((0, 2.5))
plt.ylim((0, 7))
# Setting the axes labels
plt.xlabel('m', size = 20)
plt.ylabel('b', size = 20)
# Plotting blue and red contour lines
plt.contour(M, B, Z, [10, 11, 12, 13, 14, 15, 16, 17, 18], colors = 'blue')
plt.contour(M, B, Z, [9.2, 9.3, 9.4, 9.5, 9.6, 9.7, 9.8], colors = 'red')
# Adding grid lines
plt.grid()
```
